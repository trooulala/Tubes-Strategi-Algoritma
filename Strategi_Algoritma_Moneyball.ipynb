```python
import copy
import json
```

```python
maxBudget = 68000000
```

```python
# fix Market Value format and convert Market Value type to integer 
df['Market Value'] = df['Market Value'].apply(lambda x: float(str(x).replace(',','')))
df['Market Value'] = df['Market Value'].astype(int, errors = 'raise')
```

```python
# get the GCD of Market Value
gcd = np.gcd.reduce(df['Market Value'].values)

# get the GCD of Market Value and maxBudget, we will use this value to help optimize the knapsack later
gcd = np.gcd(gcd, maxBudget)

print(f'gcd = {gcd}')
```

```python
# Create new column MV/GCD with Market Va
df['MV/GCD'] = df['Market Value'].apply(lambda x: x // gcd)
df['MV/GCD'].describe()
```

```python
formation = {
    'GK': None,
    'LB': None,
    'RB': None,
    'CBL': None,
    'CBR': None,
    'CMF': None,
    'DMFL': None,
    'DMFR': None,
    'RW': None,
    'LW': None,
    'ST': None,
}

maxBudgetGCD = maxBudget // gcd

Formations = [copy.deepcopy(formation) for _ in range(maxBudgetGCD + 1)]
Densities = [0 for _ in range(maxBudgetGCD + 1)]
Costs = [0 for _ in range(maxBudgetGCD + 1)]

# getCost = lambda formation: sum([df.loc[val]['MV/GCD'] for val in formation.values() if val])
# getDensity = lambda formation: sum([df.loc[val]['Density'] for val in formation.values() if val])

for i in range(len(df)):
    curr =  df.loc[i]

    budget = maxBudgetGCD

    while budget >= 0:
        if curr['MV/GCD'] <= budget:
            prevBudget = budget - curr['MV/GCD']

            prevFormation = Formations[prevBudget]
            
            currInRoleDensity = 0
            currInRoleCost = 0
            
            if prevFormation[curr['Role']]:
                currInRole = df.loc[prevFormation[curr['Role']]]
                currInRoleDensity = currInRole['Density']
                currInRoleCost = currInRole['MV/GCD']

            newFormation = copy.deepcopy(prevFormation)
            newFormation[curr['Role']] = i

            newDensity = Densities[prevBudget] - currInRoleDensity + curr['Density']
            newCost = Costs[prevBudget] - currInRoleCost + curr['MV/GCD']

            # newDensity = getDensity(newFormation)
            # newCost = getCost(newFormation)

            # if getDensity(Formations[newCost]) < newDensity:
            if Densities[newCost] < newDensity:
                Formations[newCost] = newFormation
                Densities[newCost] = newDensity
                Costs[newCost] = newCost

            if Densities[budget] < newDensity:
                Formations[budget] = newFormation
                Densities[budget] = newDensity
                Costs[budget] = newCost

        budget -= 1
```
